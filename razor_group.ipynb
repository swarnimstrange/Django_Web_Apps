```python
for i in range (len(df_main['businessaddress'])):
    if pd.isnull(df_main['businessaddress'].loc[i]):
        df_main['businessaddress'].loc[i] = 0
     #country.append(i[-3:])
     #print(i[-2:])
     
```

```python
df_main['country'] = df_main['businessaddress'].apply(lambda x: x[-2:] if x is not None and x != 0 else '0')
```

```python
df_main.head(7)
```

```python
df_main.drop(df_main[(df_main['country'] == 'CN')].index, inplace=True)
```

```python
df_main['country'].unique()
```

```python
len(df_main['country'].unique())
```

```python
us_ger = df_main[(df_main['country'] == 'US') | (df_main['country'] == 'DE')]
```

```python
us_ger
```

```python
us_ger.dropna(inplace = True)
```

```python
us_ger["sellerproductcount"]= us_ger["sellerproductcount"].str.split(" ", expand = False)
us_ger
```

```python
count = []
spli = us_ger["sellerproductcount"].tolist()
for i in spli:
  count.append(i[-2])
print(count)
```

```python
us_ger['sellerproductcount'] = count
us_ger.head()
```

```python
us_ger["sellerratings"]= us_ger["sellerratings"].str.split(" ", expand = False)
us_ger
```

```python
percent = []
number = []
rate = us_ger["sellerratings"].tolist()
for i in rate:
  if 'lifetime' in i:
    percent.append(i[0])
    number.append(i[-3])
  else:
     percent.append(i[0])
     number.append(i[-2])

print(percent)
print(number)
```

```python
for i in range(len(percent)):
  percent[i] = percent[i].replace('%','')
print(percent)

for i in range(len(number)):
  number[i] = number[i].replace('(','')
print(number)
```

```python
us_ger['seller ratings percent'] = percent
us_ger['seller ratings count'] = number
```

```python
us_ger_data = us_ger.copy()
```

```python
us_ger_data.drop(["sellerratings"], axis = 1, inplace = True)
us_ger_data.head()
```

```python
us_ger_data['Email_address']=us_ger_data['sellerdetails'].str.findall('(\S+@\S+)')
us_ger_data
```

```python
us_ger_data['dash_phone_number']=us_ger_data['sellerdetails'].str.extract('(\d{3}[\W]\d{3}[\W]\d{4})')
```

```python
import re
def find_phone_number(text):
    ph_no = re.findall(r"\b\d{12}\b",text)
    return "".join(ph_no)
us_ger_data['12_digit_number']=us_ger_data['sellerdetails'].apply(lambda x: find_phone_number(x))
```

```python
us_ger_data['dashed_phone_number']=us_ger_data['sellerdetails'].str.extract('(\d{5}[\W]\d{7})')
```

```python
us_ger_data.sample(10)
```

```python
num = us_ger_data["12_digit_number"].tolist()
for i in range(len(num)):
  if len(num[i]) == 24:
    num[i] = num[i][:12]
print(num)
```

```python
us_ger_data['12_digit_number'] = num
```

```python
us_ger_phone = us_ger_data[['12_digit_number', 'dash_phone_number', 'dashed_phone_number']]
```

```python
us_ger_phone
```

```python
us_ger_phone['dash_phone_number'] = us_ger_phone['dash_phone_number'].fillna(0)
us_ger_phone['dashed_phone_number'] = us_ger_phone['dashed_phone_number'].fillna(0)
```

```python
us_ger_phone['dash_phone_number'] = us_ger_phone['dash_phone_number'].apply(lambda x: x if x is not None and x != 0 else '')
us_ger_phone['dashed_phone_number'] = us_ger_phone['dashed_phone_number'].apply(lambda x: x if x is not None and x != 0 else '')
```

```python
tw_digit = us_ger_phone["12_digit_number"].tolist()
print(tw_digit)

dash_digit = us_ger_phone["dash_phone_number"].tolist()
print(dash_digit)

ger_dash = us_ger_phone["dashed_phone_number"].tolist()
print(ger_dash)
```

```python
print(len(tw_digit))
print(len(dash_digit))
print(len(ger_dash))
```

```python
phone_number = []
for i in range(len(tw_digit)):
  phone_number.append('')
print(phone_number)
```

```python
for i in range(len(phone_number)):
  if dash_digit[i] != '':
    phone_number[i] = dash_digit[i]
  if tw_digit[i] != '':
    phone_number[i] = tw_digit[i]
  if ger_dash[i] != '':
    phone_number[i] = ger_dash[i]

print(phone_number)
```

```python
us_ger_data['phone number'] = phone_number
us_ger_data.drop(["12_digit_number", "dash_phone_number", "dashed_phone_number"], axis = 1, inplace = True)
us_ger_data.head()
```

```python
us_ger_data['Email_address'] = us_ger_data['Email_address'].apply(lambda x: x[0] if x != [] else '')
```

```python
us_ger_data
```

```python
us_ger_data.drop(["sellerdetails", "businessaddress"], axis = 1, inplace = True)
```

```python
us_ger_data
```

```python
us_ger_data.to_csv('US-Germany-Data.csv', index=False)
```

# Q1 - No. of sellers in USA and Germany

```python
dataset['country'].value_counts()
```

```python
plots = sns.countplot(dataset['country'], order=dataset['country'].value_counts().index , palette='rainbow')
plt.title('Number of sellers in Germany and USA', fontsize=15)
plt.xlabel('Countries', fontsize=15)
plt.ylabel('Number of sellers', fontsize=15)


for bar in plots.patches:
    plots.annotate(format(bar.get_height()),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=14, xytext=(0, 0),
                   textcoords='offset points')
```


### Germary -: 301
### USA -: 13
